```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

# Графики зависимостей
plt.plot(x, y)


```python
x = np.linspace(0, 20, 500)
y1 = np.sin(x)
y2 = np.sin(x - np.pi)

plt.figure()

plt.plot(x, y1)
plt.plot(x, y2)
```

* `color`: Цвет линии

* `linestyle`: способ соединения точек: толстая линия, прерывистая, её отсутствие

* `linewidth`: Задаем тонкость линии

* `marker`: Что использовать для обозначения точки(круг, квадрат,треугольник, ничего).

* `markersize`: Очевидно.

* `label`: Название для данного графика в легенде.

```python
x = np.linspace(0, 20, 10000)
y1 = np.sin(x)
y2 = np.sin(x - np.pi)

plt.figure()

plt.plot(x, y1, color='gray', linestyle='-', marker='s', markersize=2, label='sinx')
plt.plot(x, y2, color='green', linestyle='--', marker='^', markersize=5, label='sin(x-pi)')
plt.legend()
```

plt.bar(left, height)

* `left`: x координата

* `height`: соответствующая высота столбца

```python
years = np.arange(2012, 2015)
values = [2, 5, 9]
plt.figure()
plt.bar(years, values)
```


* `color`: Цвет столба

* `edgecolor`: Цвет края столба

* `width`: Ширина столбика

* `label`: аналогично как с plot

```python
years = np.arange(2012, 2015)
category1_values = [2, 5, 9]
category2_values = [7, 6, 3]

plt.figure()

plt.bar(years - 0.2,
        category1_values,
        color='blue',
        edgecolor='none',
        width=0.4,
        align='center',
        label='y1')

plt.bar(years + 0.2,
        category2_values,
        color='orange',
        edgecolor='blue',
        width=0.4,
        align='center',
        label='y2')

plt.xticks(years, [str(year) for year in years])

plt.legend()
```

# Точечные диаграммы

plt.scatter(x,y)


```python
x = range(20)
y = np.arange(50, 70) + (np.random.random(20) * 10.)

plt.figure()

plt.scatter(x, y)
```

Многие функции данной библиотеки имеют похожие параметры.

```python
x = np.arange(20)
y = np.arange(50, 70) + (np.random.random(20) * 10.)

plt.figure()

plt.scatter(x, y, c='red',
            marker='s',
            edgecolor='blue')
```

# Подграфики

Позволяет вам строить несколько отдельных графиков на одном изображении.

subplot(nrows, ncols, plot_number)

* `nrows`: Количество графиков в ряд по вертикали

* `ncols`: Количество графиков в ряд по горизонтали

* `plot_number`: Расположение графика на изображении, отсчет начинается с 1.

```python
plt.figure()

for i in range(1, 7):
    plt.subplot(3, 2, i)
    plt.title(i)
    plt.xticks([])
    plt.yticks([])

plt.tight_layout()
```

# Зачем это нужно?

```python
years = np.arange(2010, 2016)

plt.figure()

for category_num in range(1, 17):
    y_vals = np.arange(10, 16) + (np.random.random(6) * category_num / 4.)
    plt.plot(years, y_vals)

plt.ylim(10, 18)
plt.xticks(years, [str(year) for year in years])
```

```python
years = np.arange(2010, 2016)

plt.figure(figsize=(10, 10))

for category_num in range(1, 17):
    plt.subplot(4, 4, category_num)
    y_vals = np.arange(10, 16) + (np.random.random(6) * category_num / 4.)
    plt.plot(years, y_vals)
    plt.ylim(10, 18)
    plt.xticks(years, [str(year) for year in years])
    plt.title('Number {}'.format(category_num))
    
plt.tight_layout()
```

# Работа с геоинформацией

По данной ссылке можно найти файл с геоинформацией


https://mapzen.com/data/metro-extracts/metro/new-york_new-york/

```python
import matplotlib.cm
from mpl_toolkits.basemap import Basemap
import pandas as pd
%time df = pd.read_csv("nyc_taxi_part.csv",index_col = "Unnamed: 0")
```

```python
plt.figure(figsize=(10,20))
_CENTER_LAT, _CENTER_LON = 40.7254, -73.86105
_LEFT_CORNER_LAT, _LEFT_CORNER_LON =  40.3528, -74.5532
_RIGHT_CORNER_LAT, _RIGHT_CORNER_LON = 41.098, -73.1689
m = Basemap( 
                    projection='cass',
                    lat_0=_CENTER_LAT, lon_0=_CENTER_LON,
                    llcrnrlon=_LEFT_CORNER_LON, llcrnrlat= _LEFT_CORNER_LAT, urcrnrlon=_RIGHT_CORNER_LON, urcrnrlat=_RIGHT_CORNER_LAT)

m.readshapefile('new-york_new-york_osm_roads_gen1', 'areas', color='green')
samples = df.sample(n=1000)
for idx in samples.index:
    x, y = m(samples['dropoff_longitude'][idx],samples['dropoff_latitude'][idx])
    m.plot(x, y, 'o', markersize=2, color='red', alpha=0.8)
```

```python
plt.figure(figsize=(10, 20))
_CENTER_LAT, _CENTER_LON = 40.7254, -73.86105
_LEFT_CORNER_LAT, _LEFT_CORNER_LON =  40.3528, -74.5532
_RIGHT_CORNER_LAT, _RIGHT_CORNER_LON = 41.098, -73.1689
m = Basemap(projection='cass',
            lat_0=_CENTER_LAT, lon_0=_CENTER_LON,
            llcrnrlon=_LEFT_CORNER_LON, llcrnrlat= _LEFT_CORNER_LAT, urcrnrlon=_RIGHT_CORNER_LON, urcrnrlat=_RIGHT_CORNER_LAT)

m.readshapefile('new-york_new-york_osm_roads_gen1', 'areas', color='green')
samples = df.sample(n=10)
for idx in samples.index:
    x, y = m(samples['dropoff_longitude'][idx], samples['dropoff_latitude'][idx])
    x1,y1 = m(samples['pickup_longitude'][idx], samples['pickup_latitude'][idx])
        # x, y = m(_CENTER_LON, _CENTER_LAT)
    m.plot([x, x1], [y, y1], 'o', markersize=2, linestyle='-', color='red', alpha=0.8)
```
